```python
import xlrd
import pandas as pd
from pandas import Series,DataFrame
import os
```

```python
path = r'.\src_data'
```

```python
if os.path.isdir(path):
    filename = path + '\\' + os.listdir(path)[0]
#filename
```

```python
wb = xlrd.open_workbook(filename)
sheets = wb.sheet_names()
sheets
```

```python
df_all = DataFrame()

for i in range(len(sheets)):
    df = pd.read_excel(filename,sheet_name = i)
    df_all = df_all.append(df)
    
print(len(df_all))
```

```python
df_all.iloc[0]
```

```python
grouped = df_all.groupby(['项目名称'
                           ,'合同编号'
                           ,'采购人'
                           ,'采购人所属地'
                           ,'项目预算'
                           ,'合同包'
                           ,'中标企业'
                           ,'中标金额'
                           ,'中标企业地址'
                           ,'中标日期'
                           ,'合同签订日期'
                           ,'合同有效期'
                           ,'交付条件'
                           ,'交付时间'
                           ,'交付地点'])
```

```python
df_grouped = grouped['支付期次','支付百分比','支付期次说明','标的','标的预算','行业'].max()
```

```python
df_grouped = df_grouped.reset_index()  #以无索引的方式返回聚合数据
```
